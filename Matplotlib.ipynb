```python
# Matplotlib workflow

# 1) Import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 2) Prepare data
a = [1, 2, 3, 4, 5]
b = [100, 150, 350, 450, 800]

# 3) Setup plot
fig, ax = plt.subplots(figsize=(10 ,10)) # width, height

# 4) Plot data
plt.plot(a, b)

# 5) Customize plot
ax.set(title='Example Plot',
       xlabel='x-axis',
       ylabel='y-axis')

# 6) Save & show plot
fig.savefig("example_plot.png")
```

## Making Figures With Numpy Array 

Note => Let's also talk about plot types.

```python
import numpy as np
```

```python
# Create some data
numpy_array = np.linspace(0, 10, 50)
numpy_array
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(numpy_array, numpy_array**2);
```

```python
# Create scatter plot with the same data
fig, ax = plt.subplots()
ax.scatter(numpy_array, np.exp(numpy_array))
ax.plot();
```

```python
# Make a scatter plot from dictionary
car_prices = {"Toyota" : 8000,
              "BMW" : 15000,
              "Opel" : 5000,
              "Mercedes" : 20000}
fig, ax = plt.subplots()
ax.scatter(car_prices.keys(), car_prices.values()) # x, y
ax.set(title="Car Sales Prices")
ax.plot();
```

```python
# Make a bar plot

print("Without list : ", car_prices.keys())
print("With list : ", list(car_prices.keys()))

# dictionary.keys() and dictionary.values() returns a generator, which matplotlib cannot use directly.

fig, ax = plt.subplots()
ax.bar(list(car_prices.keys()), list(car_prices.values())) # x, y
ax.set(title="Car Sales Prices",
       ylabel="Prices ($)",
       xlabel="Models")
ax.plot();
```

```python
# Make a horizontal bar plot

fig, ax = plt.subplots()
ax.barh(list(car_prices.keys()), list(car_prices.values())) # x, y
ax.set(title="Car Sales Prices",
       xlabel="Prices ($)",
       ylabel="Models")
ax.plot();
```

```python
# Make a histogram

random_array = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(random_array, 15) # Second parameter shows number of bins
ax.plot();
```

## Options For Subplots

```python
# Subplot Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each different axis
ax1.plot(np.array([1, 2, 3]), np.array([1, 2, 3])/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(car_prices.keys(), car_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# Plot to each different axis
ax[0, 0].plot(np.array([1, 2, 3]), np.array([1, 2, 3])/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(car_prices.keys(), car_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting From Pandas Dataframes

#### Car Sales CSV File

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales.Price = car_sales.Price.str.replace("[\$\,\.]", "")
car_sales
```

```python
# Remove last two zeros
car_sales.Price = car_sales.Price.str[:-2]
car_sales
```

```python
car_sales["Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
# cumsum() : [4000, 5000, 7000] => [4000, 4000+5000, 4000+5000+7000] (cumulative)
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum() # type(car_sales["Price"][0]) = str , we need to assign int
car_sales
```

```python
# Plotting sales
car_sales.plot(x="Date", y="Total Sales");

# Plotting scatter
car_sales["Price"] = car_sales["Price"].astype(int) # reassigning price column to int
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

#### Which one should you use? (pyplot vs matplotlib OO method)
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the Object-Oriented method

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50
```

```python
# Pyplot method
# You may externally map values to color and supply a list/array of those colors to the scatter's c argument.
over_50.plot(kind="scatter",
                     x="age",
                     y="chol",
                     c="target",
                     figsize=(10, 5));
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 5))
over_50.plot(kind="scatter",
             x="age", 
             y="chol",
             c="target",
             ax=ax);

ax.set_xlim([45, 100]); # Set the x-axis view limits (age)
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target)

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target") # Create one scatter plot per item to appear in the legend and set the label accordingly

# Add a horizontal line 
ax.axhline(over_50.chol.mean(),
           linestyle="--");
```

```python
############## Heart Disease Analysis From Scratch ##############

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Read and manipulate data
heart_disease = pd.read_csv("heart-disease.csv")
over_50 = heart_disease[heart_disease.age>50]

# Create figure two axes
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               sharex=True, # When subplots have a shared x-axis along a column, only the x tick labels of the bottom subplot are created.
                               figsize=(10, 10))

# Add data, title, meanline and legend to axes 0
ax0.scatter(over_50.age,
            over_50.chol,
            c=over_50.target,
            cmap="winter") # cmap : colormap

ax0.set(title="Hearth Disease and Cholesterol Levels",
        ylabel="Cholesterol", 
        xlim=[50, 80])

ax0.axhline(y=over_50.chol.mean(),
            color="r",
            linestyle="--")

ax0.legend(*scatter.legend_elements(), title="Target");

# Add data, title, meanline and legend to axes 1
ax1.scatter(over_50.age,
            over_50.thalach,
            c=over_50.target,
            cmap="winter") # this changes the colour scheme

ax1.set(title="Hearth Disease and Max Hearth Rate Levels",
        xlabel="Age",
        ylabel="Max Hearth Rate",
        ylim=[60, 200])

ax1.axhline(y=over_50.thalach.mean(),
            color="r",
            linestyle="--")

ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle("Hearth Disease Analysis", fontsize=15, fontweight="bold");
```
